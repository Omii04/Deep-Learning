```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt                     
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,random_state=41,n_informative=1,n_redundant=0,n_clusters_per_class=1,hypercube=False,class_sep=10)
```

```python
plt.fig_size=(10,6)
plt.scatter(X[:,0],X[:,1],c=y,cmap="winter")
```

```python
def perceptron(X,y):
    w1=w2=b=1
    lr=0.1
    
    for j in range(10000):
        for i in range(X.shape[0]):
            z=w1*X[i][0]+w2*X[i][1]+b
            
            if z*y[i]<0:
                w1=w1+lr*y[i]*X[i][0]
                w2=w2+lr*y[i]*X[i][1]
                b=b+lr*y[i]
    return w1,w2,b            
    
```

```python
w1,w2,b=perceptron(X,y)
```

```python
m=-(w1/w2)
c=-(b/w2)
print(m,c)
```
